Получить аналогичный результат на выводе:

![Необходимый результат](https://ucarecdn.com/2e821ad0-b289-47bd-bd3e-6114c79f1e2f/)

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge, Arc, PathPatch

plt.xlim(0, 6)
plt.ylim(0, 5)

ax = plt.gca()

wedge = Wedge((3, 1), 2, 45, 135)
arc = Arc((3, 1), 6, 6, 0, 45, 135, lw=3)

ax.add_patch(wedge)
ax.add_patch(arc)

plt.show()




```
