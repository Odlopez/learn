Постройте графики следующих функций на интервале от -240o до 360o:

\begin{equation}
f(x) = e^{cos(x)}+ln(sin^2(0.8x)+1)\cdot cos(x)
\end{equation}

\begin{equation}
y(x) = -ln((cos(x)+ sin(x))^2 + 1.7) + 2
\end{equation}

 

На оси OX должны быть отмечены значения в ГРАДУСАХ. 

**Пояснение.**

Для вычисления функций cos и sin значения x нужно перевести в радианы.

```python
import matplotlib.pyplot as plt
from math import cos, sin, radians, exp, log

def f_x(x):
    return exp(cos(radians(x))) + log(sin(radians(x * 0.8)) ** 2 + 1) * cos(radians(x))

def f_y(x):
    return -1 * log((cos(radians(x)) + sin(radians(x))) ** 2 + 1.7) + 2

x = [i for i in range(-240, 361)]
y_x = [f_x(i) for i in x]
y_y = [f_y(i) for i in x]

line_1 = plt.plot(x, y_x)
line_2 = plt.plot(x, y_y)

plt.setp(line_1, color='red', label='f(x)')
plt.setp(line_2, color='blue', label='f(y)')

plt.title('Графики функций f(x) и f(y)')

plt.legend()

plt.show()
```
