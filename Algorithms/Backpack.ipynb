## Задача на программирование: рюкзак

Первая строка входа содержит целые числа  $1\leq W\leq10^{4}$ и  $1\leq n\leq300$ — вместимость рюкзака и число золотых слитков. Следующая строка содержит nn целых чисел $1\leq w_{i},...w_{n} \leq10^{5}$, задающих веса слитков. Найдите максимальный вес золота, который можно унести в рюкзаке.

### Sample Input:   
10 3   
1 4 8   

### Sample Output:
9

```python
from bisect import bisect_left

W, n = [int(i) for i in input().split()]
w = sorted([int(i) for i in input().split()])

# Число, которое будет хранить максимальный вес, помещающийся в рюкзак
max_s = 0
# Заводим словарь с комбинациями веса, которые мы можем поместить в рюкзак
# Первое значение - это нулевой вес. Нулевой вес мы точно можем поместить.
dict = {0: 1}

# В отсортированном списке слитков, выбираем только те, которые точно поместятся в рюкзак, те, которые
# больше, нам просто нет смысла рассматривать
for i in range(bisect_left(sorted(w), W)):
#     Заводим массив для новых комбинаций, так как нельзя на ходу увеличивать словарь dict - 
#     питон нам не позволит этого сделать в цикле, который проходит по ключам словаря
    res = []
    # Проходим по ключам словаря, то есть по тем комбинациям, которые у нас уже есть
    for key in dict:
        # Заводим новую комбинацию, а именно - берем тот вес, который уже можно поместить в рюкзаке + вес нового слитка
        s = key + w[i]

        if s <= W:
            # Если новая комбинация меньше или равно вместимости рюкзака, до добавляем ее в массив,
            # который позже перекочует в наш словарь
            res.append(s)
            if s > max_s:
                # если новая комбинация больше, чем текущая максимальная - обновляем максимальную
                # на самом деле, если наша комбинация УЖЕ равна W - вместимости рюкзака, то мы можем прерывать проверку,
                # всё - ответ уже найден!!! Но мне лень переписывать
                max_s = s
    # добавляем новые комбинации в наш словарь
    for k in res:
        dict[k] = 1
                
print(max_s)
```
