##  Симуляция обработки сетевых пакетов

___ 

### Sample Input 1:

1 0

____ 


### Sample Output 1:

____

### Sample Input 2:

1 1   
0 0   

____ 


### Sample Output 2:

0

____ 


### Sample Input 3:

1 1   
0 1   

____ 


### Sample Output 3:

0

```python
size, n = [int(i) for i in input().split()]
packets = [[int(i) for i in input().split()] for x in range(n)]
stack = []
res = {}
end = 0

def set_el(data, el, prev_i):
    global end
  
    for it in stack:
        if res[it]['end'] <= data[0]:
            stack.remove(it)
        else:
            break

 
    if size <= len(stack):
        print(-1)
    else:
        start = end if data[0] < end else data[0]
        end = start + data[1]
        res[el] = {
          'end': end
        }
        stack.append(el)
        print(start)


for i, elem in enumerate(packets):
    set_el(elem, i, i - 1)
```
